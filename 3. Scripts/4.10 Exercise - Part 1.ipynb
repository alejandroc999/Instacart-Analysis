# #5. Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents.

```python
#Writing a function to create a customer profile flag and column
```

```python
df_filtered.columns
```

```python
def customer_profile(df_filtered):
    # Define conditions for each customer profile
    conditions = [
        (df_filtered['fam_status'] == 'married') & (df_filtered['Age'] < 35) & (df_filtered['n_dependants'] > 0),
        (df_filtered['fam_status'] == 'single') & (df_filtered['n_dependants'] > 0),
        (df_filtered['income'] > 100000) & (df_filtered['fam_status'] == 'single'),
        (df_filtered['income'] < 50000),
        (df_filtered['loyalty_flag'] == 'Loyal Customer') & (df_filtered['frequency_flag'] == 'Frequent Customer') & (df_filtered['spender_tier'] == 'High Spender'),
        (df_filtered['avg_prod_purchase_price'] < 5) & (df_filtered['n_dependants'] > 1),
        (df_filtered['fam_status'] == 'married') & (df_filtered['n_dependants'] >= 1),
        (df_filtered['median_reorder_time'] > 20) & (df_filtered['spender_tier'] > 'High Spender'),
        df_filtered['orders_day_of_week'].isin([0, 1, 6]),
        (df_filtered['avg_prod_purchase_price'] <= 5) & (df_filtered['frequency_flag'] == 'Frequent Customer')
    ]
    
    # Define corresponding customer profiles
    profiles = [
        'Young Parent',
        'Single Parent',
        'Single & High Earner',
        'Low Income Earner',
        'VIP Customer',
        'Budget Conscious Shopper',
        'Family Shopper',
        'Bulk Shopper',
        'Weekend Warrior',
        'Convenience Shopper'
    ]
    
    # Initialize the 'Customer Profile' column with a default value
    df_filtered['Customer Profile'] = 'Regular Customer'
    
    # Use np.select to assign customer profiles based on conditions
    df_filtered['Customer Profile'] = np.select(conditions, profiles, df_filtered['Customer Profile'])

# Call the function
customer_profile(df_filtered)
```

```python
#Creating a third crosstab to identify relationships between customer profiles and regions to later plot 
#Aggregate data by user ID to determine each unique user's customer profile and region, instead of each individual
#order

#Grouping the DataFrame by 'user_id' and 'Region', and calculating the median reorder time
agg_cust_data2 = df_filtered.groupby(['user_id', 'Region'])['median_reorder_time'].agg('median').reset_index()

#Define the bins for median reorder time and corresponding labels
bins = [0, 5, 10, 15, 20, 25, 30]
labels = ['0-4 days', '5-9 days', '10-14 days', '15-19 days', '20-24 days', '25-30 days']

#Grouping the 'median_reorder_time' into bins with labeled categories
agg_cust_data2['Avg Customer Reorder Time'] = pd.cut(agg_cust_data2['median_reorder_time'], bins=bins, labels=labels, right=False)

#Creating the crosstab using the labeled bins
agg_data_crosstab2 = pd.crosstab(agg_cust_data2['Region'], agg_cust_data2['Avg Customer Reorder Time'], dropna=False)

print(agg_data_crosstab2)
```

```python
stacked_data = customer_status_pricing_crosstab.div(customer_status_pricing_crosstab.sum(axis=1), axis=0)

#Get the number of bars for creating a gradient of colors
num_bars = len(df_filtered['Customer Profile'].value_counts())

#Generate a range of colors from a blue colormap, starting with a lighter shade and ending with a darker shade
colors = cm.Set3(np.linspace(0, 1, num_bars))

#Bar chart logic

plt.figure(figsize=(10, 6))
stacked_data.plot(kind='bar', stacked=True, color=colors, ax=plt.gca())
plt.title('Avg Reorder Times by Customer Profile')
plt.xticks(rotation=60)
plt.ylim(0,1)
plt.yticks([0, 1])
plt.legend(title='Avg Reorder Time Ranges', loc='upper left', bbox_to_anchor=(1, 1))
# Adding labels to the bars
for p in plt.gca().patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    plt.text(x + width / 2, y + height / 2, '{:.0%}'.format(height), 
             horizontalalignment='center', verticalalignment='center')
plt.tight_layout()
plt.savefig(os.path.join(pathname, '4. Analysis', 'Visualizations', 'median_reorder_times_stacked_by_cprofile.png'), bbox_inches='tight')
plt.show()
```
