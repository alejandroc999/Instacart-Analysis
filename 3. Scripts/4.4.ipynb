#### #Now, we will reset the strange looking index and have it say 'department_id' on the left column header and 'department' on the right column header.

#### #Now that you have an index column, let’s create a new header for your dataframe. This will be done in three steps: 1) Create a new header; 2) Remove the first row in the dataframe; and 3) Add the new header. This all sounds a little abstract, so let’s walk through the process together.

```python
# Take the first row of the df_dep_t as the new header
# iloc is a function used to assign values designated by the index number of a row or column
new_header=df_dep_t.iloc[0]
```

```python
new_header
```

```python
# Take the data under the header row for a new df
df_dep_t_new=df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# This allows for the recently created new_header variable to be used as the new column headers
df_dep_t_new.columns=new_header
```

```python
df_dep_t_new
```
