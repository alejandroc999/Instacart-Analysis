```python
#Given this large database, we will only use a subset (created here) using first million rows to avoid memory issues
df=ords_prods_merged[:1000000]
```

```python
#Using provded snippets, we will create a function with multiple conditional statements to determine product tier
# based on $$ range

def price_label(row):
    if row['prices']<=5:
        return 'Low Range Product'
    elif 5<row['prices']<=15:
        return 'Mid Range Product'
    elif row['prices']>15:
        return 'High Range Product'
    else: return 'Not enough data'
```

```python
#Here, we are going to call our recently created function (price_label) and apply it to our subset data
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Now, we check how many values of each product range there are in this subset of 1 million rows
df['price_range'].value_counts(dropna=False)
```

## #Creating new column for busiest days of the week

```python
ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

```python
#Creating an empty list that will be inputed when creating the values for a new column in the ords_prods_merge df
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Running the result list
result
```

```python
#Now, we create the new column in the ords_prods_merged df called "busiest" days and will input the result list
ords_prods_merged['busiest_day']=result
```

```python
#Using the value.counts() function to check the frequency of each value
ords_prods_merged['busiest_day'].value_counts(dropna=False)
```

## #Creating an altered 'results' list and 'busiest_day' column

```python
#Step 1
#Creating an empty list that will be inputed when creating the values for a new column in the ords_prods_merge df.
#The conditional if statements are categorizing values from the 'orders_day_of_week' column
result1 = []

for value in ords_prods_merged['order_day_of_week']:
  if value == 0 or value==1:
    result1.append('Busiest day')
  elif value == 4 or value==3:
    result1.append('Least busy')
  else:
    result1.append('Regularly busy')
```

```python
#Step 2
#Creating a new column 'busiest_days' in the ords_prods_merger dataframe which uses ccreated 'result1' values
ords_prods_merged['busiest_days']=result1
```

```python
#Using the value.counts() function to check the frequency of each value in new 'result1' list
ords_prods_merged['busiest_days'].value_counts(dropna=False)
```

```python
#Step 1
#Creating an empty list that will be inputed when creating the values for a new column in the ords_prods_merge df.
#The conditional if-statements are categorizing values from the 'order_hour_of_day' column 
result2 = []

for value in ords_prods_merged['order_hour_of_day']:
  if value in [10, 11, 14, 15]:
    result2.append('Most orders')
  elif value in [3, 4, 2, 5]:
    result2.append('Least Orders')
  else:
    result2.append('Average_orders')
```

```python
#Step 2
#Creating a new column 'busiest_hours' in the ords_prods_merger dataframe which categorizes 'result2' values
ords_prods_merged['busiest_hours']=result2
```

```python
#Checking the value frequency for the newly created 'busiest_hours' column
ords_prods_merged['busiest_hours'].value_counts(dropna=False)
```
