### #6. Create a spending flag for each user based on the average price across all their orders using the following criteria:

### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#First step: We will group by 'user_id' and find the mean 'prices' value for every customer on the database. 
ords_prods_merge.groupby('user_id')['prices'].mean()
```

```python
#Now, we will create a new column in the dataframe which contains every users average product purchase price.
ords_prods_merge['avg_prod_purchase_price']=ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
#Now, like done multiple times above, we shall create a new column called 'spender_tier' and two different categories
#in which we classify Instacart users. First, we will do 'Low Spender', classified as avg_prod_purchase_price<10
ords_prods_merge.loc[ords_prods_merge['avg_prod_purchase_price']<10, 'spender_tier']='Low Spender'
```

```python
#Now, we will create the 'High Spender' tier, classified as avg_prod_purchase_price>=10
ords_prods_merge.loc[ords_prods_merge['avg_prod_purchase_price']>=10, 'spender_tier']='High Spender'
```

```python
#Now, we will test by using the value_counts() func to see value frequency in the new column
ords_prods_merge['spender_tier'].value_counts(dropna=False)
```

### #7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
#First step: We will group by 'user_id' and find the median 'days_since_last_order' value for every customer on
#the database.
ords_prods_merge.groupby('user_id')['days_since_last_order'].median()
```

```python
#Now, we will create a new column in the dataframe which contains every users median amt of days until reordering
ords_prods_merge['median_reorder_time']=ords_prods_merge.groupby('user_id')['days_since_last_order'].transform(np.median)
```

```python
#Now, like done multiple times above, we shall create a new column called 'frequency_flag' and three different 
#categories in which we classify Instacart users. First, we will do 'Non Frequent Customer'
#classified as median_reorder_time>20
ords_prods_merge.loc[ords_prods_merge['median_reorder_time']>20, 'frequency_flag']='Non Frequent Customer'
```

```python
#Now, we will create the 'Regular Customer' tier
#Classified as 10<median_reorder_time>=20
ords_prods_merge.loc[(ords_prods_merge['median_reorder_time']>10)&(ords_prods_merge['median_reorder_time']<=20), 'frequency_flag']='Regular Customer'
```

```python
#Finally, we will create the 'Frequent Customer' tier
#Classified as median_reorder_time<10
ords_prods_merge.loc[ords_prods_merge['median_reorder_time']<=10, 'frequency_flag']='Frequent Customer'
```

```python
#Now, we will test by using the value_counts() func to see value frequency in the new column
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
