```python
#First, we are going to write similar code to previous exercises to identify if there are any columns with mixed
#data types with a for loop and if statement, finding it thru the assignment of Boolean values to determine whether
#or not the condition is met
for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

# #6.Combine customer data with revised Instacart data

```python
##We could or should have done this in the first section, but we will now import the Instacart dataframe in order
##to merge it with the customer set
ords_prods_revised=pd.read_pickle(os.path.join(pathname, '2. Data', 'Prepared Data', 'ords_prods_merge_final.pkl'))
```

```python
#In order to have the same data types for the key column, we will need to change the data type for the 'user_id'
#column in the ords_prods_final dataframe to match the object data type we applied to the same column in the 
#customer 'df' dataframe
ords_prods_revised['user_id']=ords_prods_revised['user_id'].astype('str')
```

```python
##We will combine this customer dataframe with the revised ords_prods_merge dataframe containing the majority of 
##information on Instacart data
instacart_final=df.merge(ords_prods_revised, on='user_id', indicator=True)
```
